```python
import numpy as np
import pandas as pd
import datetime
```

```python
def _df_rename_team(df, tar_col):
    df = df.reset_index(drop = True).copy()
    new_name_list = []
    M_TEAM_NAMES = pd.read_csv('data/other/TEAM_NAMES.csv').values.tolist()
    
    for i, row in df.iterrows():
        count_change = 0
        for i, team_list in enumerate(M_TEAM_NAMES):
            if row[tar_col] in team_list:
                new_name_list += [team_list[0]]
                count_change += 1
        if count_change == 1:
            pass
        elif count_change == 0:
            new_name_list += [row[tar_col]]
            print(f"TEAM_NAMES.csvに記載なし：{row[tar_col]}")
        else:
            print(f"TEAM_NAMES.csvを修正してください：{row[tar_col]}")
            
    df.drop(columns = [tar_col], inplace = True)

    df_tmp = pd.DataFrame(new_name_list, columns = [tar_col])
        
    df = pd.concat([df, df_tmp],axis=1)
    return df
```

```python
def get_last_team(player, df_marge):
    columns1 = ['年月日', 'H_Team', 'A_Team']
    columns2 = ['H_選手1', 'H_選手2', 'H_選手3', 'H_選手4', 'H_選手5', 'H_選手6', 'H_選手7', 'H_選手8', 'H_選手9', 'H_選手10', 'H_選手11']
    columns3 = ['A_選手1', 'A_選手2', 'A_選手3', 'A_選手4', 'A_選手5', 'A_選手6', 'A_選手7', 'A_選手8', 'A_選手9', 'A_選手10', 'A_選手11']
    df = df_marge.copy()
    
    df = df[columns1 + columns2 + columns3]
    last_play_match = df[(df == player).sum(axis='columns').astype(bool)].sort_values('年月日', ascending=False)[:1]
    
    return last_play_match
    
    if (last_play_match[columns2]==player).sum(axis='columns').astype(bool).values[0]:
        last_team = last_play_match['H_Team'].values[0]
    else:
        last_team = last_play_match['A_Team'].values[0]
    return last_team
```

```python
# 読み込み
df_player_list = pd.read_csv('data/j_league_data_site/j_player_list_cleaned.csv', index_col=0).reset_index(drop = True).copy()
df_marge = pd.read_csv('data/marge/marge.csv', index_col=0).dropna().reset_index(drop = True).copy()
```

```python
df_player_list = _df_rename_team(df_player_list, '最終所属')
df_player_list['key'] = df_player_list['選手名'] + df_player_list['最終所属']
df_player_list['P_ID'] = df_player_list.index
```

```python

for head in ['H_', 'A_']:
    for i in range(1, 12):
        p_id_list = []
        for j, row in df_marge.iterrows():
            player = row[head + '選手' + str(i)]
            last_team = get_last_team(player, df_marge)
            
            try:
                p_id = df_player_list[df_player_list['key']==player+last_team]['P_ID'].values[0]
                p_id_list += [p_id]
            except IndexError:
                # Jリーグ公式HPに更新漏れの情報があるため、その対応。。。
                if player == '大内　一生' and last_team = 'Ｙ．Ｓ．Ｃ．Ｃ．横浜':
                    p_id = df_player_list[df_player_list['key']=='大内　一生'+'横浜ＦＣ']['P_ID'].values[0]
                    p_id_list += [p_id]
                
            
        df_tmp = pd.DataFrame(p_id_list, columns = [head + '選手' + str(i) + '_ID'])
        df_marge = pd.concat([df_marge, df_tmp],axis=1)
```
